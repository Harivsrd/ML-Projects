```python
import pandas as pd
data = pd.read_csv("students.csv")
print(data)
```

```python
print(data.head())
print(data.describe())
print(data.info())
```

```python
import matplotlib.pyplot as plt

data.plot(kind="bar")
plt.show()
```

```python
data["Average"] = (data["Math"] + data["Physics"] + data["Chemistry"]) / 3

print(data)
```

```python
data["Result"] = data["Average"].apply(lambda x: "Pass" if x > 60 else "Fail")

print(data)
```
